```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('poe_stats.csv')
df.head()
```

```python
df.info()
```

```python
df['ladder'].value_counts()
```


Grafik PIE


```python
fig, ax = plt.subplots(figsize=(8,4))
ax.pie(df['ladder'].value_counts(),startangle=90, autopct='%0.1f%%')
ax.axis('equal')
ax.legend(['SSF Harbinger','SSF Harbinger HC','Harbinger','Hardcore Harbinger'], title='Ladder', loc='lower left')
plt.ylabel('Ladder List', size=14)
plt.xlabel('Presentase Ladder', size=14)
plt.xticks(size=8)
plt.yticks(size=12)
ax.set_title('Grafik Ladder')
plt.show()
```

GRAFIK BATANG

```python
plt.figure(figsize=(12,7))
plt.bar(['SSF Harbinger','SSF Harbinger HC','Harbinger','Hardcore Harbinger'], df['ladder'].value_counts())

plt.title('Grafik Batang Ladder', size=16)
plt.ylabel('Jumlah Ladder', size=14)
plt.xlabel('Nama Ladder', size=14)
plt.xticks(size=8)
plt.yticks(size=12)

plt.show()
```

GRAFIK PLOT

```python
plt.figure(figsize=(10,5))
plt.plot(df['ladder'].value_counts(), ['SSF Harbinger','SSF Harbinger HC','Harbinger','Hardcore Harbinger'], label ='Garis')
plt.legend()
plt.title('Grafik Plot', size=16)
plt.ylabel('Nama Ladder', size=14)
plt.xlabel('Jumlah Ladder', size=14)
plt.xticks(size=8)
plt.yticks(size=12)
plt.show()
```

GRAFIK SCATTER PLOT

```python

sns.scatterplot(df['ladder'])
plt.title('Grafik Scatter Plot', size=16)
plt.ylabel('Nama Ladder', size=14)
plt.xlabel('Jumlah Ladder', size=14)
plt.xticks(size=8)
plt.yticks(size=12)
plt.show()
```
